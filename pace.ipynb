```python
import fastf1
from fastf1 import plotting
from matplotlib import pyplot as plt
plotting.setup_mpl()
```

```python
fastf1.Cache.enable_cache('cache')  # optional but recommended

session = fastf1.get_testing_session(2022, 2, 1)
laps = session.load_laps(with_telemetry=True).pick_accurate()

driver1 = laps.pick_driver('LEC')
driver2 = laps.pick_driver('HAM')
driver3 = laps.pick_driver('NOR')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(driver1['LapNumber'], driver1['LapTime'], color='red', label='Leclerc')
ax.plot(driver2['LapNumber'], driver2['LapTime'], color='cyan', label='Hamilton')
ax.plot(driver3['LapNumber'], driver3['LapTime'], color='orange', label='Norris')
ax.set_title("")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.legend(loc='upper right', prop={"size":20})
plt.show()
```
